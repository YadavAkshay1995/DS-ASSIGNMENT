```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movies = pd.read_csv(r"C:\Users\Akshay Yadav\Downloads\Association-Rules-movies-dataset--main\Association-Rules-movies-dataset--main\my_movies.csv")
movies.head(10)
```

```python
new_movies = movies.iloc[:,5:]
new_movies
```

```python
frequent_itemsets = apriori(new_movies, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 0.7)
rules
```

```python
rules[rules.lift>1]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,7))
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('SUPPORT VS CONFIDENCE')
plt.show()
```
